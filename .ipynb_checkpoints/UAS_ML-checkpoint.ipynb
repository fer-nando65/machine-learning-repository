```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

### Importing The Dataset

```python
df = pd.read_csv('csv_result-dataset_24_mushroom.csv')
df.head()
```

```python
df.info()
```

### Cleaning The Dataset

```python
df.drop('id', axis=1, inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

### Normalisize The Dataset

```python
le = LabelEncoder()

for i in df.columns:
    df[i] = le.fit_transform(df[i])

df.head()
```

### Splitting The Dataset into Train and Test

```python
X = df.drop("'class'", axis=1)
y = df["'class'"]
```

```python
seed = 48
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1, random_state=seed)
```

### Create Decision Tree Classifier Model

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=5)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
auc_train = metrics.roc_auc_score(y_train, y_train_pred)
auc_test = metrics.roc_auc_score(y_test, y_test_pred)

print("AUC train = {}\nAUC test = {}".format(round(auc_train, 4), round(auc_test, 4)))
```

```python
print(classification_report(y_test, y_test_pred))
```

### Plotting The Model

```python
var_columns = [c for c in X.columns]

plt.figure(figsize=(30, 10))

plot_tree(model,
         feature_names=var_columns,
         class_names=['E', 'P'],
         rounded=True,
         filled=True,
         fontsize=10)

plt.show()
```
